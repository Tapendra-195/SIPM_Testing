```python
import pandas as pd
import matplotlib.pyplot as plt

chans = [1,2,3,5]
dfs=[]
for chan in chans:
    dfs.append( pd.read_csv('test_ch'+str(chan)+'.csv',skipinitialspace=True) )
            
dfs[1].keys()
```

```python
plt.figure( figsize=(3,3), dpi=100)
for i,chan in enumerate(chans):
    plt.plot( dfs[i]['HV_set (V)'], dfs[i]['HV_rdb (V)']+i, '.', label='CH'+str(chan)+'+'+str(i))
plt.ylabel('HV_rdb (V)')
plt.xlabel('HV_set (V)')
plt.legend()
plt.tight_layout()
plt.savefig('HVset-rdb-ch1-4.png',facecolor='w')
```

```python
plt.figure( figsize=(3,3), dpi=100)
for i,chan in enumerate(chans):
    plt.plot( dfs[i]['HV_set (V)'], dfs[i]['HV_err (V)'], '.', label='CH'+str(chan))

plt.ylabel('HV_err (V)')
plt.xlabel('HV_set (V)')
plt.legend()
plt.tight_layout()
plt.savefig('HVerr-uwpg01-ch1-4.png',facecolor='w')
```

```python
# 'VsCH_2 (V)', 'IsCH_2 (A)'
plt.figure( figsize=(3,3), dpi=100)
for i,chan in enumerate(chans):
    plt.plot( dfs[i]['VsCH_'+str(chan)+' (V)'], dfs[i]['IsCH_'+str(chan)+' (A)'],  '.', label='CH'+str(chan) )
plt.xlabel('Vs (V)')
plt.ylabel('Is (A)')
plt.legend()
plt.tight_layout()
plt.savefig('IV-curve-uwpg01-ch1-4.png',facecolor='w')
```
